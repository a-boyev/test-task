## 1. Задача на базовый питон (математика)

Напишите функцию, которая будет суммировать только целые И нечётные значения.


```python
new_list = [0,3.2,3,15,13,1]

def summator(lst):
    s = 0
    for i in lst:
        if (int(i) == i) and (i % 2 == 1): s = s + i
    return s

print(summator(new_list))
```

## 2. Задача на базовый питон (логическое мышление)
Напишите функцию, которая будет рассчитывать угол между часовой и минутной стрелкой в заданное время.

```python
def delta(hours,minutes):
    hours_angle = hours * 30
    minutes_angle = minutes * 6
    diff = max(hours_angle, minutes_angle) - min(hours_angle, minutes_angle)
    return min(diff, 360 - diff) 

print(delta(6.5, 30))
```

## 3. Машинное обучение
Во вложении приложен файл с датасетом с именем "test_data.csv". В нём расположены данные по работе условной станции,а именно:
* timestamp_hour_end - время события;
* PBR - плановый график работы по заданию Системного оператора, МВтч;
* power - фактическая выработка электроэнергии, МВтч;
* pressure_msl -атмосферное давление, , измеряемая в единицах давления земной атмосферы на уровне моря (1013 гектопаскалей)
* relativehumidity_2m - относительная влажность на уровне 2м от уровня земли, %;
* temperature_2m -температура атмосферного воздуха на уровне 2м от уровня земли, град цельсия;
* winddirection_10m - направление ветра на уровне 10м от уровня земли, градусы;
* windspeed_10m - скорость ветра на уровне 10м от уровня земли, м/с.

Целевым признаком для прогноза на 2 суток вперёд (+48 часов) является столбец "power".

Необходимо проанализировать данные, отфильтровать данные для обучения, обучить минимум 3 типами моделей и сравнить данные.

Задача оценивается по умению работы с данными и их подготовки к обучению, генерации признаков, интерпретации результатов (в том числе и с помощью визуальных инструментов).

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn import linear_model
from math import sqrt
import os, math
import numpy as np

# Cleanup data (remove NA and zero values)
df = pd.read_csv("test_data.csv")
df = df.dropna()
df = df.drop(df[df['PBR'] == 0].index)
df = df.drop(df[df['power'] == 0].index)

len_forecast = 48 # Length of forecast interval
block_size = len_forecast * 3
len_dataset = len(df)
num_blocks = int(len_dataset / block_size)
train_test_ratio = 4
predict_key = 'power'
       
# Prepare dataset (split test/train, add extra features like wind_x, wind_y)
def prepare_dataset(df):
    train_x = []
    train_y = []
    test_x = []
    test_y = []
    
    train_columns = df.columns.drop(predict_key).drop('timestamp_hour_end').drop('winddirection_10m').drop('windspeed_10m')
    def get_features(idx):
        features = []
        row = df.iloc[idx]
        dt = datetime.strptime(row['timestamp_hour_end'], '%Y-%m-%d %H:%M:%S')
        features.append(dt.month)
        features.append(dt.day)
        features.append(dt.hour)
        wind_x = row.windspeed_10m * math.cos(row.winddirection_10m / 90 * 1.57)
        wind_y = row.windspeed_10m * math.sin(row.winddirection_10m / 90 * 1.57)
        features.append(wind_x)
        features.append(wind_y)
        features.extend(list(row[train_columns]))
        return features
    
    for i in range(num_blocks):
        if i % 10 == 0: print("Progress %f %%" % (i / num_blocks * 100))
        for j in range(len_forecast):
            features = []
            idx = i * block_size + j
            for k in range(len_forecast):
                features.extend(get_features(idx + k))
            if i % train_test_ratio != 0:
                train_x.append(features)
                train_y.append(df.iloc[idx + 2 * len_forecast][predict_key])
            else:
                test_x.append(features)
                test_y.append(df.iloc[idx + 2 * len_forecast][predict_key])
    return train_x, train_y, test_x, test_y

print("Preparing dataset")
train_x, train_y, test_x, test_y = prepare_dataset(df)
print("Preparing dataset complete")

# Type 1 - SVM regression

print("Training SVM regressor")
res = SVR().fit(train_x, train_y)
predictions_svm = res.predict(test_x)
rmse_svm = sqrt(mean_squared_error(test_y, predictions_svm))
print("Training SVM regressor complete")
print("SVM regressor RMSE value %f : " % rmse_svm)

plt.plot(test_y)
plt.plot(predictions_svm)
plt.show()

# Type 2 - Multi-layer perceptron

print("Training MLP regression")
mlp = MLPRegressor().fit(train_x, train_y)
predictions_mlp = mlp.predict(test_x)
rmse_mlp = sqrt(mean_squared_error(test_y, predictions_mlp))
print("Training MLP regression complete")
print("SVM regressor RMSE value %f : " % rmse_mlp)

plt.plot(test_y)
plt.plot(predictions_mlp)
plt.show()

# Type 3 - Bayesian Ridge regression
print("Training Bayesian Ridge regression")
res_bayes = linear_model.BayesianRidge().fit(train_x, train_y)
predictions_bayes = res_bayes.predict(test_x)
rmse_bayes = sqrt(mean_squared_error(test_y, predictions_bayes))
print("Training Bayesian Ridge regression complete")
print("SVM regressor RMSE value %f : " % rmse_bayes)

plt.plot(test_y)
plt.plot(predictions_bayes)
plt.show()
```

## 4. Междисциплинарная задача

Есть узел, в который подходит три трубы, как на рисунке.

Приборы на трубах показывают, что в узел поступает 90, по левой трубе уходит 35, а по правой - 60. Чего именно - не важно, это может быть электрический ток, расход воды и т.п. Важно, что в узле нет ни потерь, ни источников.

Вопрос состоит в том, сколько на самом деле течет по каждой из труб.




![task4](./3Pipes.png)

```python
import scipy.optimize as optimize
from scipy import stats
import numpy as np

def calc(x_in, x_out_l, x_out_r, num_attempts = 10):
    def eval_step(params):
        mu_in, mu_out_l, std = params
        prob_in = stats.norm.pdf(x_in, mu_in, std)
        prob_out_l = stats.norm.pdf(x_out_l, mu_out_l, std)
        prob_out_r = stats.norm.pdf(x_out_r, mu_in - mu_out_l, std)
        return - np.log(prob_in) - np.log(prob_out_l) - np.log(prob_out_r)
    min_val = np.inf
    min_params = None
    for i in range(num_attempts):
        mu_in = np.random.normal(x_in, 5)
        mu_out_l = np.random.normal(x_out_l, 5)
        std = np.random.uniform(low=1, high=5)
        initial_guess = [mu_in, mu_out_l, std]
        result = optimize.minimize(eval_step, initial_guess, bounds=((60, 110), (10, 40), (0, 20))).x
        prob = eval_step(result)
        if prob < min_val:
            min_val = prob
            min_params = result
    return min_params

mu_in, mu_out_l, std = calc(90, 35, 60)
print("in: %f, out_left: %f, out_right: %f" % (mu_in, mu_out_l, mu_in - mu_out_l))
```
